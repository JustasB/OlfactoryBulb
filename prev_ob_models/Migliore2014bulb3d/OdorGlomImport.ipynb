```python
with open('input-odors.txt') as f:
    text = f.read()
```

```python
rows = text.split('\r\n')
rows = [row.split('\t') for row in rows]
rows = rows[:-1]
rows = [[col.replace(' ','') for col in row] for row in rows]
rows = [[float(col) if c > 0 else col for c, col in enumerate(row)] for row in rows]
```

```python
for row_id, row in enumerate(rows):
    odor_id = row_id + 1
    odor = row[0]

    # Intensities start at col 2
    # Each column corresponds to Mig glom index
    for mig_glom_id, intensity in enumerate(row[1:]): 
        birg_glom_id = mig2birg[mig_glom_id]
        #print(odor, odor_id, mig_glom_id, birg_glom_id, intensity)
        print('insert into odor_glom (odor_id, glom_id, intensity) values (%s, %s, %s);'%(odor_id, birg_glom_id, intensity))
        
        
```
