```python
import os
files = os.listdir(dir)

maxlines = None
for filename in files:
    with open(dir + "\\" + filename) as f:
        lines = f.readlines()
        if maxlines is None or maxlines < len(lines):
            maxlines = len(lines)
            print(filename + str(len(lines)))
            
```

```python
maxlines
```

```python
sholls = []
for filename in files:
    with open(dir + "\\" + filename) as f:
        lines = f.readlines()
        radii = []
        intersections = []
        for line in lines:
            line = line.split(',')
            if line[0] == "Radius":
                continue
            print(line)
            radii.append(int(line[0]))
            intersections.append(float(line[1]))
        #intersections = list(np.array(intersections)/np.max(intersections))
        while len(radii) < maxlines:
            radii.append(radii[-1]+10)
            intersections.append(0)
        sholls.append([filename,radii,intersections])
```

```python
i = -10
for i in range(-30,-20):
    plt.plot(sholls[i][1],sholls[i][2])
    plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
X = np.array([sholl[2] for sholl in sholls])
```

```python
pca = PCA(n_components=.90,svd_solver='full',whiten=False)
pca.fit(X)
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
T = pca.transform(X)
#T = X
```

```python
T
```

```python
plt.scatter(list(zip(*T))[0],list(zip(*T))[1])
```

```python
import matplotlib.pyplot as plt  
import pandas as pd  
import numpy as np 
```

```python
from sklearn.cluster import AgglomerativeClustering
n_clusters=3
cluster = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')  
cluster.fit_predict(T)  

for i, lab in enumerate(cluster.labels_):
    if len(sholls[i]) < 4:
        sholls[i].append(lab)
    else:
        sholls[i][3] = lab
```

```python
plt.scatter(T[:,0],T[:,1], c=cluster.labels_, cmap='rainbow')  
```

```python
def cluster_sholl(cluster=None):
    data = []

    for i,bin in enumerate(sholls[0][1]):
        box = []
        for sholl in sholls:
            if cluster==None or cluster == sholl[3]:
                box.append(sholl[2][i])
        data.append(box)

    plt.figure(figsize=(10, 7))
    plt.plot([np.median(bin) for bin in data])
    plt.plot([np.quantile(bin, .95) for bin in data])
    plt.plot([np.quantile(bin, .05) for bin in data])
    plt.show()
```

```python
%matplotlib inline
cluster_sholl()
cluster_sholl(0)
cluster_sholl(1)
cluster_sholl(2)
```
