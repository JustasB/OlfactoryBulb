```python
def fill_test_observations(cell_type):
    no_obs_tests = []
    for test in soma_tests + basal_tests + apical_tests:
        try:
            test.observation = next({'mean':p.mean*test.units, 'std':p.std*test.units, 'n':p.n} for p in Property.select(Property).where(Property.id == get_property_id(cell_type,test.__class__.__name__)))
        except TypeError:
            no_obs_tests.append(test)
            
    return no_obs_tests
```

```python
def judge_cell_models(cell_type):
    pubs = eval(cell_type.lower()+"_pubs")
    no_obs_tests = fill_test_observations(cell_type)

    tests = soma_tests + basal_tests + apical_tests
    
    # remove missing obs tests
    tests = [test for test in tests if test not in no_obs_tests]
    
    full_suite = TestSuite(tests)
    
    models = []
    for pub in pubs:
        dir = 'prev_ob_models/'+pub+'/morphology/'
        for file in os.listdir(dir):
            if cell_type.lower() in file and '.swc' in file:
                model_file = 'prev_ob_models/'+pub+'/morphology/'+file
                models.append(SwcCellModel(model_file, name=pub + ' ' + file))
        
    
    return full_suite, full_suite.judge(models)
```

```python
def get_all_cell_type_swcs(cell_type):
    swc_list = "morphology-data/"+cell_type+"s.txt"
    with open(swc_list,'r') as f:
        swcs = f.readlines()
        swcs = ["morphology-data" + f.strip()[1:] for f in swcs]
    return swcs
```

```python
def judge_cell_type_swcs(cell_type):
    swcs = get_all_cell_type_swcs(cell_type)
    tests = soma_tests + basal_tests + apical_tests
    full_suite = TestSuite(tests)

    models = [SwcCellModel(file,name=file) for file in swcs]
    score = full_suite.judge(models)
    return score
```
